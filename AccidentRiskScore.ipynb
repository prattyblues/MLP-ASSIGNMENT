```python
x=df.drop(["AccidentRiskIndex"],axis=1)
y=df["AccidentRiskIndex"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 0,test_size=0.25)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_train)
```

```python
print("R squared: {}".format(r2_score(y_true=y_train,y_pred=y_pred)))
```

```python
residuals = y_train.values-y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```

```python
#Mean of residual is close to 0, so we are good here.
```

```python
p = sns.scatterplot(y_pred,residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(residuals, X_train)
lzip(name, test)
```

```python
#Since p value is more than 0.05 in Goldfeld Quandt Test, we can't reject it's null hypothesis that error terms are homoscedastic.
```

```python
p = sns.distplot(residuals,kde=True)
p = plt.title('Accident Risk Index Score')
```

```python
#The residual terms aren't normally distributed for the number of test points we took. 
```

```python
from scipy.stats import bartlett
test = bartlett(X_train[0],residuals)
print(test)

test = bartlett(X_train[1],residuals)
print(test)

test = bartlett(X_train[2],residuals)
print(test)
```

```python
#p-value that corresponds to the test statistic is less than some significance level then we can reject the null hypothesis and cannot conclude that not all groups have the same variance.
#Not good for regression
```

```python
plt.figure(figsize=(10,5))
p = sns.lineplot(y_pred,residuals,marker='o',color='blue')
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='red')
p = plt.title('Residuals vs fitted values plot for autocorrelation check')
```

```python
from statsmodels.stats import diagnostic as diag
min(diag.acorr_ljungbox(residuals , lags = 40)[1])
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(residuals, lags=40)
plt.show()
```

```python
sm.graphics.tsa.plot_pacf(residuals, lags=40)
plt.show()
```

```python
#The results do not show signs of autocorelation since there are no spikes outside the blue confidence interval region
```
